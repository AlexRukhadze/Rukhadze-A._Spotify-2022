###  Краткое описание датасета
**"Spotify Top Chart Songs 2022"** содержит данные о 646 песнях, которые "вирусились" на стриминговом сервисе Spotify на сентябрь 2022 года. Датасет состоит из 17 переменных, среди них есть как текстовые (название трека и имя исполнителя), так и количественные переменные (примеры: пиковая позиция трека, число недель в чарте, продолжительность трека и др.). Подробное описание переменных датасета представлено ниже.

### Описание переменных датасета
• **uri** - URI-код трека<br>
• **artist_names** - Имя (имена) исполнителя (исполнителей) трека<br>
• **track_name** - Название трека<br>
• **peak_rank** - Пиковая позиция трека в чарте Spotify<br>
• **weeks_on_chart** - Число недель, проведенных в чарте Spotify<br>
• **danceability** - Степень танцевальности трека (от 0.0 до 1.0, где 0.0 - нетанцевальная, а 1.0 - танцевальная)<br>
• **energy** - Степень энергичности трека (от 0.0 до 1.0, где 0.0 - неэнергичная, а 1.0 - энергичная)<br>
• **key** - Тональность трека (от 0 до 11, где 0 - до-мажор, 1 = до-диез мажор/ре-бемоль минор и т.д.)<br>
• **loudness** - Громкость трека (в dB)<br>
• **mode** - Лад трека (0 или 1, где 0 - минор, а 1 - мажор)<br>
• **speechiness** - Преобладание слов в треке (от 0.0 до 1.0, где 0.0 - только музыка, а 1.0 - только слова)<br>
• **acousticness** - Акустичность трека (от 0.0 до 1.0, где 0.0 - неакустическая, а 1.0 - акустическая)<br>
• **instrumentalness** - Инструментальность трека (от 0.0 до 1.0, где 0.0 - вокальная песня, а 1.0 - инструментальная)<br>
• **liveness** - Форма исполнения трека (от 0.0 до 1.0, где 0.0 - песня записана в студии, а 1.0 - песня исполнена вживую перед зрителями)<br> 
• **tempo** - Темп трека (в BPM; чем выше значение, тем быстрее темп песни)<br>
• **time_signature** - Размер такта в треке (от 3 до 7, где 3 - это 3/4, 4 - это 4/4 и т.д.)<br> 
• **duration_ms** - Продолжительность трека (в мс, миллисекундах)

**Вопрос 4.3**: Как получить новую переменную с категориями треков по их темпу (*'tempo'*)?

```python
spotify['tempo'].mean()
```

Средний темп треков в датасете составляет *~120 BPM*, поэтому будет отталкиваться от этого значения и разделим все треки по темпу на следующие четыре категории:

```python
def get_tempo_cat (tempo):
    if tempo < 90:
        return 'Low'
    elif tempo < 120:
        return 'Middle-'
    elif tempo < 150:
        return 'Middle+'
    elif tempo >= 150:
        return 'High'
    else:
        return tempo
```

```python
spotify['tempo_cat'] = spotify['tempo'].apply(get_tempo_cat)
```

```python
spotify['tempo_cat'] = pd.Categorical(spotify['tempo_cat'])
```

```python
spotify.head()
```

**Вывод 4.3**: Через **def get_tempo_cat** задаем четыре категории, на которые разбиваем значения переменной *'tempo'*. Новую переменную *'tempo_cat'* мы переводим в категориальный тип через **pd.Categorical( )** и сохраняем в наш датасет.

**Вопрос 6.7**: Какие значения разбросов по длительности треков и числу недель пребывания в чарте Spotify в 2022 году (*'duration_sec' и 'weeks_on_chart'*) в зависимости от переменной *'chart_top'*, если считать по интерквартильному размаху, дисперсии и стандартному отклонению?

```python
def range_values(x):
    return x.max() - x.min()

spotify.pivot_table(index='chart_top', values=['weeks_on_chart', 'duration_sec'], aggfunc=['var', 'std', range_values])
```

**Вывод 6.7**: Для ответа на этот вопрос нам требуется построить пивотальную таблицу через **pivot.table( )**. В целом, можно отметить, что более сильна видна разница в разбросах по дисперсии (**var**) и стандартному отклонению (**std**), нежели чем по размаху (**range_values**), где разница по *'duration_sec'* составляет всего 7 секунд, а по *'weeks_on_chart'* - 3 недели.

**Вопрос 9.3**: Как выглядит линейный график суммарного числа недель пребывания треков в чарте в зависимости от тональности по каждой категории темпа?

```python
weeks_sum = spotify.groupby(['tempo_cat', 'key'])['weeks_on_chart'].sum()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 15))

ax[0][0].plot(weeks_sum.loc['Low'], color = '#FECBFA', linewidth=3, marker = 's', markersize=8)
ax[0][1].plot(weeks_sum.loc['Middle-'], color = '#1DB954', linewidth=3, marker = 's', markersize=8)
ax[1][0].plot(weeks_sum.loc['Middle+'], color = '#B8FB60', linewidth=3, marker = 's', markersize=8)
ax[1][1].plot(weeks_sum.loc['High'], color = '#D1D1D1', linewidth=3, marker = 's', markersize=8)

for axis in ax.ravel():
    axis.set_ylim(0, 1250)
    axis.set_xlabel('Тональность трека (key)')
    axis.set_ylabel('Число недель пребывания трека в чарте (weeks_on_chart)')
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)
    
ax[0][0].legend(['Низкий темп (Low tempo)'], frameon=False)
ax[0][1].legend(['Темп ниже среднего (Middle- tempo)'], frameon=False)
ax[1][0].legend(['Темп выше среднего (Middle+ tempo)'], frameon=False)
ax[1][1].legend(['Высокий темп (High tempo)'], frameon=False)

fig.suptitle('Суммарное число недель пребывания треков в чарте Spotify', fontsize=15)
```

**Вывод 9.3**: Линейный график (с объединением данных), построенный через **plt**, состоит из 4 под-графиков, каждый из которых соответствует определенной категории темпа трека (*переменная 'tempo_cat'*). На каждом под-графике по оси X мы видим изменение тональности трека (*переменная 'key'*), а по оси Y - суммарное число недель пребывания треков в чарте Spotify (*'weeks_sum'*). Можно отметить, что дольше всего в чарте держатся песни с темпом ниже среднего ('Middle-'), исполненные в тональностях, зашифрованных под значениями 1 и 8.
